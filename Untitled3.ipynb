```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
onRet = pd.read_excel('Online Retail.xlsx')
```

```python
onRet.shape
```

```python
train_set = onRet.iloc[:500]
```

```python
train_set.shape
```

```python
#loc is like iloc but works with boolean. Basically, if the tuple isn't null, we incorporate it into the DF
nulles_train = train_set.loc[pd.isnull(train_set.CustomerID) == False]
```

```python
nulles_train.shape
```

```python
train_set.head()
```

```python
items = nulles_train[['StockCode', 'Description']].drop_duplicates()
items['StockCode'] = items.StockCode.astype(str)
```

```python
items.head()
```

```python
#just making sure customerId is int types and not string or double or whatever
nulles_train['CustomerID'] = nulles_train.CustomerID.astype(int)
nulles_train.head()
```

```python
#We want to keep only the columns we'll need. Invoice and country info is not required
#description is already in our item DF
nulles_train = nulles_train[['StockCode', 'Quantity', 'CustomerID']]
```

```python
nulles_train.head()
```

```python
#now group
grouped = nulles_train.groupby(['CustomerID', 'StockCode'])['Quantity'].sum().reset_index()
#remember we deleted rows, so index are unpair
#sum() method: total # of an item x by customerID
```

```python
grouped.head()
```

```python
grouped.Quantity.loc[grouped.Quantity == 0] == 1
```

```python
purchased = grouped.query('Quantity > 0')
#http://jose-coto.com/query-method-pandas
```

```python
purchased
```

```python
no_products = len(purchased.StockCode.unique())
no_products
```

```python
no_customers = len(purchased.CustomerID.unique())
no_customers
```

```python
rating = purchased.pivot(index = 'CustomerID', columns = 'StockCode', values = 'Quantity').fillna(0).astype(int)
```

```python
rating
```

```python
bin_rat = rating.copy()
```

```python
bin_rat[bin_rat != 0] = 1
```

```python
bin_rat
```

```python
prods_int = np.zeros((no_products,no_products))
```

```python
for i in range(no_products):
    for j in range(no_products):
        if i!=j:
            DF_ij = bin_rat.iloc[:,[i,j]]
            sum_ij = DF_ij.sum(axis=1)
            pair_ij = len(sum_ij[sum_ij == 2])
            prods_int[i,j] = pair_ij
            prods_int[j,i] = pair_ij
```

```python
tims_pur = prods_int.sum(axis=1)
```

```python
tims_pur
```

```python
prods_weighted = np.zeros((no_products,no_products))
for i in range(no_products):
    for j in range(no_products):
        if(tims_pur[i]+tims_pur[j]) !=0:
            prods_weighted[i,j] = (prods_int[i,j])/(tims_pur[i]+tims_pur[j])
```

```python
nodes_codes = np.array(bin_rat.columns).astype('str')
item_dict = pd.Series(items.Description.values, index = items.StockCode).to_dict()
node_labels = [item_dict[code] for code in nodes_codes]
```

```python
import networkx as nx
```

```python
G = nx.from_numpy_matrix(prods_weighted)
pos=nx.random_layout(G)
labels = {}
for idx, node in enumerate(G.nodes()):
    labels[node] = node_labels[idx]

nx.draw_networkx_nodes(G, pos , node_color="skyblue", node_size=30)
nx.draw_networkx_edges(G, pos,  edge_color='k', width= 0.3, alpha= 0.5)
nx.draw_networkx_labels(G, pos, labels, font_size=4)
plt.axis('off')
plt.show()
```

```python
#Export graph to Gephi
H=nx.relabel_nodes(G,labels) 
nx.write_gexf(H, "products.gexf")
```

```python
from community import community_louvain
partition = community_louvain.best_partition(G)
values = list(partition.values())
```

```python
len(np.unique(values))
```

```python
prod_comm = pd.DataFrame(node_labels, columns = ['Product_description'])
prod_comm['CommunityID'] = values
```

```python
prod_comm[prod_comm['CommunityID']==14].head(15)
```

```python
weight_prod_DF = pd.DataFrame(prods_weighted, columns = node_labels)
weight_prod_DF.set_index(weight_prod_DF.columns, 'product', inplace=True)

prod_prob = weight_prod_DF.divide(weight_prod_DF.max(axis=1), axis=0)
```
